```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv")

df.head()
```

```python
df.dtypes
```

```python
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12,8))
sns.lineplot(data=df, x="reportStartDateTime", y="dryBulbTemperature_Celsius")
plt.xticks(rotation=90)
plt.title("Dry Bulb Temperature")
plt.xlabel("Datetime")
plt.ylabel("Temperature (C)")
plt.show()
```
